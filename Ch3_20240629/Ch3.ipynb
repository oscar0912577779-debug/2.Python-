# Conditional Filtering

````python
df.head()
````

````python
bool_series=df["total_bill"]>30
df[bool_series].head()
````

````python
df[df["total_bill"]>30].head()
````

````python
df[df['sex']=='Male'][0:5]
````

````python
df[df['sex']=='Male'].iloc[0:5]
````

````python
df[  (df["total_bill"]>30)   &   (df["sex"]=="Male")  ].head()
````

````python
df[   (df["day"]=="Sun")     |    (df["day"]=="Sat")  ].head()  #| shift+\
````

````python
b1=True
b2=False
print(b1 or b2)
b1=True
b2=False
print(b1 | b2)
b1=1
b2=0
print(b1 and b2)
b1=1
b2=0
print(b1 & b2)
````
